```python
import pandas as pd

chemin = 'resultats-elections-presidentielles-2022-1er-tour.csv'

# Chargement du fichier
fichier = pd.read_csv(chemin)
```

```python

fichier.head() # Test #
```

```python
# Importe numpy pour la fonction abs
import numpy as np


colonnes_quantitatives = fichier.select_dtypes(include=[np.number])

# Affiche les 5 premières lignes
colonnes_quantitatives.head()
```

```python
# Calcul des statistiques s
moyennes = colonnes_quantitatives.mean().round(2).tolist()
medianes = colonnes_quantitatives.median().round(2).tolist()
modes = colonnes_quantitatives.mode().iloc[0].round(2).tolist()
ecarts_type = colonnes_quantitatives.std().round(2).tolist()

# Calcul de l'écart absolu à la moyenne
ecarts_absolus_moyens = np.abs(colonnes_quantitatives - colonnes_quantitatives.mean()).mean().round(2).tolist()

# Calcul de l'étendue
etendues = (colonnes_quantitatives.max() - colonnes_quantitatives.min()).round(2).tolist()
```

```python

noms_colonnes = colonnes_quantitatives.columns.tolist()

print("--- Noms des colonnes numériques ---")
print(noms_colonnes)
print("\n--- Statistiques (Arrondies à 2 décimales) ---")

print("Moyennes :", moyennes)
print("Médianes :", medianes)
print("Modes :", modes)
print("Écarts Types :", ecarts_type)
print("Écarts Absolus à la Moyenne :", ecarts_absolus_moyens)
print("Étendues (Max - Min) :", etendues)
```

```python

import numpy as np
import pandas as pd


# Calcul de la Distance Interquartile

q1 = colonnes_quantitatives.quantile(0.25)
q3 = colonnes_quantitatives.quantile(0.75)
distance_interquartile = (q3 - q1).round(2).tolist()

# Calcul de la Distance Interdécile
d1 = colonnes_quantitatives.quantile(0.10)
d9 = colonnes_quantitatives.quantile(0.90)
distance_interdecile = (d9 - d1).round(2).tolist()

print("\n--- Distances Interquartiles et Interdéciles (Arrondies à 2 décimales) ---")
print("Distance Interquartile (IQR) :", distance_interquartile)
print("Distance Interdécile (IDR) :", distance_interdecile)
```

```python
import matplotlib.pyplot as plt
import os


output_folder = 'img'
if not os.path.exists(output_folder):
    os.makedirs(output_folder)
    print(f"Dossier '{output_folder}' créé.")


for colonne in colonnes_quantitatives.columns:
    plt.figure(figsize=(8, 6))


    colonnes_quantitatives.boxplot(column=colonne)

    plt.title(f'Boîte à Moustache pour la colonne : {colonne}')
    plt.ylabel('Valeur')
    plt.grid(False)

 # Sauvegarde
    file_path = os.path.join(output_folder, f'boxplot_{colonne}.png')
    plt.savefig(file_path)
    plt.close() # Ferme la figure pour libérer la mémoire

print(f"\nBoîtes à moustaches générées et sauvegardées dans le dossier '{output_folder}'.")
```
